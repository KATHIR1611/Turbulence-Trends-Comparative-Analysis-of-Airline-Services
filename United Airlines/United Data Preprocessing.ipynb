```
import pandas as pd
from datetime import datetime
```

```
df=pd.read_csv("/work/Raw_United_Airlines_reviews.csv")
df
```

```
import re
def clean_date(x):
    return re.sub(r'(\d+)(st|nd|th|rd)',r'\1',x)
df.Date=df.Date.apply(clean_date)
df.Date=pd.to_datetime(df.Date)
```

```
df.Date=pd.to_datetime(df.Date).dt.date
```

```
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")

lemma=WordNetLemmatizer()
stop_words=set(stopwords.words("english"))

reviews_data=df.Reviews.str.replace("✅ Trip Verified |","",regex=False)
reviews_data=df.Reviews.str.replace("Not Verified |","",regex=False)
reviews=[]

for rev in reviews_data:
    rev=rev.lower()
    rev=rev.split()
    rev=[lemma.lemmatize(word) for word in rev if word.isalpha() and word not in stop_words]
    rev=" ".join(rev)
    reviews.append(rev)
```

```
df=df.drop(columns=["Reviews"])
```

```
df["Review"]=reviews
```

```
df["Review"]=df["Review"].str.replace("trip verified","",regex=False)
```

```
df.head()
```

```
na=df.pop("Review")
df.insert(0,"Review",na)
df
```
