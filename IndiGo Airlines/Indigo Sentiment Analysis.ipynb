```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df=pd.read_csv("/work/Indigo_Data.csv")
```

```
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
vader=SentimentIntensityAnalyzer()
```

```
def get_vader_sentiment(review):
    sentiment_scores=vader.polarity_scores(review)
    if sentiment_scores["compound"]>= 0.05:
        return "Positive"
    elif sentiment_scores["compound"]<= -0.05:
        return "Negative"
    else:
        return "Neutral"
```

```
df["Vader_Polarity"]=df["Review"].apply(get_vader_sentiment)
df
```

```
from transformers import pipeline
analyzer=pipeline("sentiment-analysis")

def get_huggingface(review):
    result=analyzer(review)
    return result[0]["label"]
```

```
df["HF_Label"]=df["Review"].apply(get_huggingface)
df
```

```
df["HF_Label"].value_counts()
```

```
df["VADER_Polarity"].value_counts()
```

```
df["HF_Label"]=df["HF_Label"].replace({"NEGATIVE":"Negative","POSITIVE":"Positive"})
df
```

```
df.rename(columns={"Vader_Polarity":"VADER_Polarity"},inplace=True)
df
```

```
a=df[["VADER_Polarity","HF_Label"]].apply(pd.Series.value_counts,axis=0)
a.plot(kind="bar",stacked=True,figsize=(6,6),colormap="viridis", edgecolor="black")
plt.title("VADER vs Hugging Face Sentiment Analysis")
plt.xlabel("Sentiment Types")
plt.ylabel("No. of Reviews ")
plt.show()
```
