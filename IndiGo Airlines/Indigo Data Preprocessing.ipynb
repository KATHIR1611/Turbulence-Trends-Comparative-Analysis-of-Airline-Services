```
df["Stars"].fillna(df["Stars"].mean(),inplace=True)
```

```
df["Stars"]=df["Stars"].astype(int)
```

```
df.info()
```

```
import re
def clean_date(x):
    return re.sub(r'(\d+)(st|nd|rd|th)',r'\1',x)
df["Date"]=df["Date"].apply(clean_date)
df["Date"]=pd.to_datetime(df["Date"])
```

```
from datetime import datetime
df["Date"]=pd.to_datetime(df["Date"]).dt.date
```

```

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("omw-1.4")

lemma=WordNetLemmatizer()
stop_words=set(stopwords.words("english"))

reviews_data=df.Review.str.replace("✅ Trip Verified |","",regex=False)
reviews_data=df.Review.str.replace("Not Verified |","",regex=False)
Reviews=[]

for rev in reviews_data:
    rev=rev.lower()
    rev=rev.split()
    rev=[lemma.lemmatize(word) for word in rev if word.isalpha() and word not in stop_words]
    rev=" ".join(rev)
    Reviews.append(rev)
```

```
df["Reviews"]=Reviews
df
```

```
df=df.drop(columns=["Review"])
df
```

```
na=df.pop("Reviews")
df.insert(0,"Reviews",na)
df
```

```
df.rename(columns={"Reviews":"Review"},inplace=True)
df
```
